```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')

df.head()
```

```python
# Inspección General
df.shape
df.info()
df.describe(include='all')
```

```python
# Limpieza
df.isnull().sum()
df.nunique()
```

```python
# Gráficos con Plotly Express
df = df[(df['price'] > 0) & (df['price'] < 10000)]
    # Histograma de precios
px.histogram(df, x='price', nbins=50, title='Distribución de precios')
    
    # Histograma de año del modelo
px.histogram(df, x='model_year', title='Año del modelo de los vehículos')

    # Grafico de disperción: precio vs kilometraje
px.scatter(df, x='odometer', y='price', title='Precio vs Kilometraje', trendline='ols')

    # Boxplot: precio por tipo de vehículo
px.box(df, x='type', y='price', title='Precio por tipo de vehículo')
```

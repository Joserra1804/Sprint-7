```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')

df.head()
```

```python
# Inspecci√≥n General
df.shape
df.info()
df.describe(include='all')
```

```python
# Limpieza
df.isnull().sum()
df.nunique()
```

```python
# Gr√°ficos con Plotly Express
df = df[(df['price'] > 0) & (df['price'] < 10000)]
    # Histograma de precios
px.histogram(df, x='price', nbins=50, title='Distribuci√≥n de precios')
    
    # Histograma de a√±o del modelo
px.histogram(df, x='model_year', title='A√±o del modelo de los veh√≠culos')

    # Grafico de disperci√≥n: precio vs kilometraje
px.scatter(df, x='odometer', y='price', title='Precio vs Kilometraje', trendline='ols')

    # Boxplot: precio por tipo de veh√≠culo
px.box(df, x='type', y='price', title='Precio por tipo de veh√≠culo')
```
